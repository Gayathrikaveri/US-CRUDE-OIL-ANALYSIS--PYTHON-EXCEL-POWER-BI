```python
 # Getting outliers using Tukey method (IQR method)

# Calculate Q1 and Q3
Q1 = df['quantity'].quantile(0.0)
Q3 = df['quantity'].quantile(0.506)

# Calculate IQR
IQR = Q3 - Q1


# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Get outliers
outliers= df[(df['quantity'] < lower_bound) | (df['quantity'] > upper_bound)]

# Display outliers count
print('Total Outliers detected:',len(outliers))
```

What happens to that dataset if outliers are removed?

```python
#records left if outliers are removed from the dataset
print('Before removing outliers:',len(df))
print('After removing outliers:',len(df)-len(outliers))
#percentage of data loss if outliers are removed
print('Percentage of data loss:',round((1-((len(df)-len(outliers))/len(df)))*100,2),"%")
```

1) Trimming method 

```python
# Make a copy of the original DataFrame
df_copy = df.copy()

# Calculate Q1 and Q3
Q1 = df_copy['quantity'].quantile(0.0)
Q3 = df_copy['quantity'].quantile(0.506)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the copy
df_no_outliers = df_copy[(df_copy['quantity'] >= lower_bound) & (df_copy['quantity'] <= upper_bound)]

# Count of records removed
print('No.of records removed:',len(df)-len(df_no_outliers))

# Display records left
print('Total records left:',len(df_no_outliers))
```

```python
df_no_outliers['quantity'].describe()
```

2)With Capping (Without removal of outliers)

Capping: In capping, outliers are adjusted by replacing them with a predefined upper value (cap). This technique ensures that outliers are limited to a maximum acceptable value.

```python
# Make a copy of the original DataFrame
df_cap=df.copy()

Q1 = df_cap['quantity'].quantile(0.0)
Q3 = df_cap['quantity'].quantile(0.55)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cap.loc[(df_cap['quantity']>upper_bound),'quantity']=upper_bound
df_cap.loc[(df_cap['quantity']<lower_bound),'quantity']=lower_bound

# Display records left
print('Total records left:',len(df_cap))
```

```python
df_cap['quantity'].describe()
```

3) Flooring (Without removal of outliers)

Flooring: In flooring, outliers are adjusted by replacing them with a predefined lower value (floor). This technique ensures that outliers are not completely removed but rather set to a minimum acceptable value.

```python
# Make a copy of the original DataFrame
df_floor = df.copy()

# Define the floor threshold (lower bound)
floor_threshold = 1  # Set the threshold value above which outliers will be floored

# Floor outliers
df_floor.loc[df_floor['quantity'] > floor_threshold, 'quantity'] = floor_threshold

# Display outliers count
print('Total Outliers detected:', (df_floor['quantity'] > floor_threshold).sum())

# Display records left
print('Total records left:',len(df_floor))
```

```python
df_floor['quantity'].describe()
```

4)Tranformation using log

When data is highly skewed or has a long tail, applying a logarithmic transformation can help make the distribution more symmetric and easier to analyze.

```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame containing the 'quantity' column

# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = df['quantity'].quantile(0.0)
Q3 = df['quantity'].quantile(0.55)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# # Create a mask to identify outliers
outliers_mask = (df['quantity'] < lower_bound) | (df['quantity'] > upper_bound)

# # Create a new column for log-transformed outliers
df['quantity_log'] = df['quantity']

# # Apply log transformation to outliers
df.loc[outliers_mask, 'quantity_log'] = np.log(df.loc[outliers_mask, 'quantity'])
```

```python
df['quantity_log'].describe()
```

5)WINSORZATION

Is a data transformation technique used to mitigate the impact of extreme values (outliers) in a dataset by replacing them with less extreme values within a specified range, typically determined by percentiles. It helps stabilize statistical measures by reducing the influence of outliers while preserving the overall distribution of the data.

```python
import numpy as np
new_win=df.copy()

def winsorize(data, lower_percentile=0, upper_percentile=88):
    # Calculate percentiles
    lower_bound = np.percentile(data, lower_percentile)
    upper_bound = np.percentile(data, upper_percentile)
    
    # Winsorize the data
    winsorized_data = np.where(data < lower_bound, lower_bound,
                               np.where(data > upper_bound, upper_bound, data))
    
    return winsorized_data
new_win['quantity']=winsorize(new_win['quantity'])
```

```python
new_win['quantity'].describe()
```

```python
# Load the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

country_counts = df[df['originTypeName'] == 'Country']['originName'].value_counts()

# Get the top 10 origin with the highest counts
top_10_countries = country_counts.head(10)

# Filter the world dataframe to include only the geometries for the top 10 origin
highlighted_countries = world[world['name'].isin(top_10_countries.index)]

# Plot the world map
world.plot(color='lightgrey', edgecolor='white', figsize=(20, 15))

# Plot the highlighted origin
highlighted_countries.plot(color='lightgreen', ax=plt.gca())

# Annotate each origin with its count for the top 10 countries
for idx, row in highlighted_countries.iterrows():
    country_name = row['name']
    count = top_10_countries.get(country_name, 0)
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, 
             f"{country_name}: ", 
             ha='right', fontsize=12, color='blue', fontweight='bold')  # Country name in blue and bold
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, 
             f"{count}", 
             ha='left', fontsize=12, color='darkgreen', fontweight='bold')
plt.title('Top 10 countries that export highest Crude oil to US')
# Show the plot
plt.show()
```
